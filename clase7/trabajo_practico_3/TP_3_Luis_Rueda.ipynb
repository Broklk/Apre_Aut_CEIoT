### 1. Modelado y Evaluación

Se selecciona el modelo de bosques aleatorios

```python
# Importamos librerías necesarias
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import balanced_accuracy_score, f1_score
```

```python
# Cargamos los datos

df = pd.read_csv('weatherAUS_limpio.csv')
x_test = pd.read_csv('x_test.csv')
y_test = pd.read_csv('y_test.csv')
x_train = pd.read_csv('x_train.csv')
y_train = pd.read_csv('y_train.csv')
```

```python
# Implementación del modelo

random_forest_classifier = RandomForestClassifier(random_state = 42)
random_forest_classifier.fit(x_train, y_train)
```

```python
# Realizamos predicciones

y_pred = random_forest_classifier.predict(x_test)
```

#### 1.1. Evalúe el modelo con las métricas definidas.

```python
# Calculamos métricas de evaluación (hiperparámetros por defecto)

balanced_accuracy = balanced_accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precisión Balanceada: {balanced_accuracy*100:.2f}%")
print(f"F1-score: {f1*100:.2f}%")
```

#### 1.2. Discuta las fortalezas y debilidades del modelo elegido en el contexto específico de este problema.

##### Fortalezas:

- A diferencia de un solo árbol de decisión, los bosques aleatorios son menos propensos a sobreajustes a los datos de entrenamiento debido a la agregación de resultados de multiples árboles. Por lo cual, los bosques aleatorios son mas robustos en la predicción de lluvia que un solo árbol.
- Los bosques aleatorios son buenos para relaciones complejas entre distintas variables, como la humedad, temperatura y presión.
- Los bosques aleatorios pueden manejar conjuntos de datos grandes y de alta dimensionalidad, lo que es beneficioso si se dispone de datos meteorológicos variados.

##### Debilidades:

- Los árboles de decisión individuales son fáciles de interpretar, los bosques aleatorios, que consisten en cientos o miles de árboles, son mucho más difíciles de interpretar.
- Entrenar un modelo de Random Forest suele ser más costoso en términos de tiempo y recursos computacionales, especialmente cuando se trabaja con un gran número de árboles y un gran conjunto de datos extenso.
- Aunque los bosques aleatorios son robustos, el rendimiento puede depender de una buena selección de hiperparámetros. Si estos parámetros no se ajustan correctamente, el modelo puede no funcionar de manera óptima.
